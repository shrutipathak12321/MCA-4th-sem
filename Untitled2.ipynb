```python
import cv2
import numpy as np

def threshold_segmentation(image):
    # Define intensity thresholds for white matter, gray matter, and CSF
    wm_lower, wm_upper = 200, 255  # Adjust as needed
    gm_lower, gm_upper = 100, 200  # Adjust as needed
    csf_lower, csf_upper = 0, 100  # Adjust as needed

    # Apply thresholding to segment white matter, gray matter, and CSF
    wm_mask = cv2.inRange(image, wm_lower, wm_upper)
    gm_mask = cv2.inRange(image, gm_lower, gm_upper)
    csf_mask = cv2.inRange(image, csf_lower, csf_upper)

    # Create background mask (complement of all other masks)
    background_mask = cv2.bitwise_not(cv2.bitwise_or(wm_mask, gm_mask, csf_mask))

    # Combine masks to create segmented image
    segmented_image = np.zeros_like(image)
    segmented_image[wm_mask > 0] = 255  # White matter
    segmented_image[gm_mask > 0] = 128  # Gray matter
    segmented_image[csf_mask > 0] = 64   # CSF
    segmented_image[background_mask > 0] = 0  # Background

    return segmented_image

# Load ground truth image
ground_truth_image = cv2.imread("slice_0.jpg", cv2.IMREAD_GRAYSCALE)

# Apply thresholding to segment the ground truth image
segmented_image = threshold_segmentation(ground_truth_image)

# Display the segmented image
cv2.imshow("Segmented Image", segmented_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Save the segmented image
cv2.imwrite("segmented_image.jpg", segmented_image)
```

```python
import os
from sklearn.model_selection import train_test_split

# Function to split dataset into train, validation, and test sets
def split_dataset(image_dir, gt_dir, test_size=0.2, val_size=0.1, random_state=42):
    # Get list of image filenames
    image_filenames = sorted(os.listdir(image_dir))
    
    # Get corresponding list of ground truth filenames
    gt_filenames = sorted(os.listdir(gt_dir))
    
    # Split filenames into train and test sets
    image_filenames_train, image_filenames_test, gt_filenames_train, gt_filenames_test = train_test_split(
        image_filenames, gt_filenames, test_size=test_size, random_state=random_state)
    
    # Split train filenames into train and validation sets
    image_filenames_train, image_filenames_val, gt_filenames_train, gt_filenames_val = train_test_split(
        image_filenames_train, gt_filenames_train, test_size=val_size/(1-test_size), random_state=random_state)
    
    return (image_filenames_train, gt_filenames_train), (image_filenames_val, gt_filenames_val), (image_filenames_test, gt_filenames_test)

# Example usage:
image_dir = r"E:\Shruti\Extracted_files\pgm\t1_icbm_normal_1mm_pn1_rf20"
gt_dir = r"E:\Shruti\Extracted_files\pgm\phantom_1.0mm_normal_crisp"
test_size = 0.2  # 20% of the data will be used for testing
val_size = 0.1  # 10% of the data will be used for validation
random_state = 42  # Random seed for reproducibility

# Split dataset
(train_files, train_gt_files), (val_files, val_gt_files), (test_files, test_gt_files) = split_dataset(image_dir, gt_dir, test_size, val_size, random_state)

# Store file paths in x_train, y_train, x_val, y_val, x_test, y_test
X_train = [os.path.join(image_dir, filename) for filename in train_files]
y_train = [os.path.join(gt_dir, filename) for filename in train_gt_files]

X_val = [os.path.join(image_dir, filename) for filename in val_files]
y_val = [os.path.join(gt_dir, filename) for filename in val_gt_files]

X_test = [os.path.join(image_dir, filename) for filename in test_files]
y_test = [os.path.join(gt_dir, filename) for filename in test_gt_files]

# Display the number of files in each split
print("Number of training files:", len(X_train))
print("Number of validation files:", len(X_val))
print("Number of test files:", len(X_test))
```
