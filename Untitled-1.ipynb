```python
import os

def read_pgm(filename):
    """
    Reads a PGM file and returns the image data as a 2D list.
    """
    with open(filename, 'rb') as f:
        # Skip comments
        magic_number = f.readline().strip()
        if magic_number != b'P2':
            raise ValueError("Not a PGM file")
        while True:
            line = f.readline().decode('utf-8').strip()
            if not line.startswith('#'):
                break

        # Read the header
        width, height = map(int, line.split())
        max_val = int(f.readline())

        # Read the image data
        pixel_values = []
        for _ in range(height):
            row = map(int, f.readline().split())
            pixel_values.extend(row)

        return pixel_values
    
filename_mask='Extracted_files\\pgm\\phantom_1.0mm_normal_crisp\\slice_80.pgm'
pixel_values_mask = read_pgm(filename_mask)
```

```python
print(pixel_values_mask)
```

```python
height = 217
width = 181
```

```python
def convert_to_2d_list(lst, rows, cols):
    if len(lst) != rows * cols:
        raise ValueError("Length of the 1D list does not match the specified dimensions of the 2D list.")

    return [lst[i*cols:(i+1)*cols] for i in range(rows)]


img = convert_to_2d_list(pixel_values_mask, height, width)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 4))  # Set the size of the figure
plt.axis('off')  # Turn off the axis
plt.title('Total')
plt.imshow(img)  # Display the image in grayscale
plt.savefig('total.png', bbox_inches='tight')  # Save the plot as an image file without extra whitespace
plt.show()

```

```python
from PIL import Image
import os

def crop_image(image_path, crop_box):
    with Image.open(image_path) as img:
        cropped_image = img.crop(crop_box)
        cropped_image.save(image_path)

def crop_images_in_folder(folder, crop_box):
    """
    Crops all images in the specified folder and saves them to the same folder.
    
    :param folder: Folder containing images to be cropped.
    :param crop_box: A tuple (left, upper, right, lower) defining the crop box.
    """
    temp_folder = os.path.join(folder, 'temp')
    if not os.path.exists(temp_folder):
        os.makedirs(temp_folder)

    # Move original images to the temporary folder
    for filename in os.listdir(folder):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            original_path = os.path.join(folder, filename)
            temp_path = os.path.join(temp_folder, filename)
            os.rename(original_path, temp_path)
    
    # Process images from the temporary folder and save back to the original folder
    for filename in os.listdir(temp_folder):
        temp_path = os.path.join(temp_folder, filename)
        original_path = os.path.join(folder, filename)
        crop_image(temp_path, crop_box)
        os.rename(temp_path, original_path)

    # Remove the temporary folder
    os.rmdir(temp_folder)

# Example usage
# folder = f'E:\Shruti\segmented_image'
crop_box = (60, 80, 590, 390)  # Adjust these values as needed

crop_images_in_folder(folder, crop_box)

```
